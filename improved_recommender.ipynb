```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
similarities = cosine_similarity(vectors)
```

```python
new_df["title"] = new_df["title"].str.strip()
```

```python
def recommend(movie):
    # movie = movie.strip().lower()

    if movie not in new_df['title'].values:
        return f"'{movie}' not found in the database. Please try another movie."

    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarities[index])), reverse=True, key=lambda x: x[1])

    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```

```python
def suggest_by_genre(genre):
    genre = genre.lower()
    filtered = movies[movies['genres'].apply(lambda genres: genre in [g.lower() for g in genres])]
    filtered = filtered.copy()

    return filtered.sort_values(by = ['score'], ascending = False)[['title']].head(5)
```

```python
suggest_by_genre("Comedy")
```

```python
def suggest_by_genre(genres_list):
    genres_list = [g.lower() for g in genres_list]

    def match_all(genre_tags):
        movie_genres = [tag.lower() for tag in genre_tags]
        return all(g in movie_genres for g in genres_list)

    filtered = movies[movies['genres'].apply(match_all)].copy()

    if filtered.empty:
        return f"No movies found matching genres: {genres_list}"

    return filtered.sort_values(by = ['score'], ascending = False)[['title']].head(5)
```

```python
suggest_by_genre(["Action", "Comedy"])
```
